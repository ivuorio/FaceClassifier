# Deep Learning Classifier

```python

import numpy as np
import os
from skimage import io
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
```

## Reading the image datas from folder

```python
#Variable for the folder path
folderPath = '../Untitled Folder/yalefaces_crop/'
#list for file directories to be read
dataPath = []
#read individual files
for i in os.listdir(folderPath):
    if i != 'Readme.txt':
        dataPath.append(os.path.join(folderPath, i))
        
#list for the images and labels        
imData = []
imLabels = []

#loop for all files that were found on the given folder
for file in dataPath:
    #read single image
    imRead = io.imread(file, as_grey=True)
    #append to a list of images 
    imData.append(imRead)
    #read the label of the image
    labelRead = int(os.path.split(file)[1].split("B")[1].split("_")[0]) 
    #append the label to a list of labels
    imLabels.append(labelRead)
```

## Checking data to be correct
Check the amount of images, number of label and print one image.

```python
#Number of pictures
amount = len(imData)
#random image to be printed
x = imData[np.random.randint(0,amount)]
#plot image wiht mathplotlib
plt.imshow(x, cmap='gray')
#number of labels
labels = len(np.unique(imLabels))
print ('Number of images ', amount)
print ('Number of unique labels ', labels)
plt.show()
```

## Reorganising the data

```python
x=np.array(imData)
y = np.array(imLabels)
#amount of data used for training
trnSize = 0.2
x_train, x_test, y_train, y_test = train_test_split(x, y, 
    train_size=trnSize)
```

## Modifying data for the training alorithm 


```python
#change the data types to be float32 numbers instead of integers
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

#make values to be from between 0 and 1
x_train /= 255.
x_test /= 255.

# reshape to: numberOfSamples, channels, height, width
x_train = x_train.reshape(x_train.shape[0], 1, x_train.shape[1], x_train.shape[2])
x_test = x_test.reshape(x_test.shape[0], 1, x_test.shape[1], x_test.shape[2])

#print the shapes to check that all is right
print("Training matrix shape", x_train.shape)
print("Testing matrix shape", x_test.shape)
```

## Building the Neural network

```python
import tensorflow as tf
from keras.models import Model
from keras.models import Sequential
from keras.layers import Convolution2D, MaxPooling2D, Flatten, Dense
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, History, Callback


def convMLC():
    sizeX = x_train.shape[2]
    sizeY = x_train.shape[3]
    #choosing the model
    model = Sequential()
    
    ############ADDING LAYERS###########
    
    #First Block
    model.add(Convolution2D(16, (5, 5), activation='relu', name='blk1_conv1', input_shape=(1, sizeX, sizeY),
                            strides=(1, 1), padding="same", kernel_initializer="he_normal"))
    model.add(MaxPooling2D((2, 2), strides=(2, 2), name="blk1_pool", data_format="channels_first"))
    
    #Second Block
    model.add(Convolution2D(26, (5, 5), activation='relu', name='blk2_conv1',
                            strides=(1, 1), padding="same", kernel_initializer="he_normal"))
    model.add(MaxPooling2D((2, 2), strides=(2, 2), name="blk2_pool", data_format="channels_first"))
    model.add(Flatten())

    #Third Block
    model.add(Dense(512, activation="relu", name="fc1", kernel_initializer="he_normal"))
    model.add(Dropout(rate=0.5))

    model.add(Dense(40, activation="softmax", name="prediction", kernel_initializer="he_normal"))
    optim = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
    
    model.compile(loss='sparse_categorical_crossentropy', optimizer=optim, metrics=['accuracy'])
    
    return model
model = convMLC()
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, to_file='model.png')
```

## Training the NN

```python
from time import time
import keras.backend as ks
#save starting time
t0=time()
#set a parameter for loss history
loss_history = History()

ks.set_value(model.optimizer.lr,0.0005)
#train the model
model.fit(x_train, y_train, batch_size=64, epochs=100, 
          verbose=1, validation_data=(x_test, y_test), 
          callbacks=[loss_history])
#the time it took to train
timeFit = time()-t0
```

## Plotting of the training history

```python
import pandas as pd

# Plot the training curve of the model
df = pd.DataFrame(loss_history.history)
df.loc[:,'Iter'] = np.arange(1, len(loss_history.history['acc'])+1, 1)
plt.plot(df.Iter,df.acc, label='Accuracy at training data')
plt.plot(df.Iter,df.val_acc, label='Accuracy at testing data')
plt.legend()
plt.savefig('acc%0.2f.png' %trnSize, dpi=199)
plt.show()
```

```python
#plot the curve of cost of the model 
plt.plot(df.Iter, df.loss, label='Value of cost function')
plt.plot(df.Iter, df.val_loss, label='Val. of cost func. for cross validation data')
plt.legend()
plt.savefig('loss%0.2f.png' %trnSize, dpi=199)
plt.show()
```

```python
#calculate the time it takes to predict the test set values
t0 = time()
y_pred = model.predict(x_test)
timePred = time()-t0
```

## Testing the quality of the classifier

```python
#pick the most propable prediction as the prediction.
#yp for the prediction
yp = []
for i in y_pred:
    yp.append(np.argmax(i))
print('Confusion Matrix')
print(confusion_matrix(y_test, yp))
```

```python
#Plot the confusion matrix
plt.imshow(confusion_matrix(y_test, yp, labels=range(labels+1)));
plt.colorbar()
plt.savefig('confNN%0.2f.png' %trnSize, dpi=199)
plt.show()
```

```python
print('Total time for fit: %0.2f'%(timeFit))
print('Total time for prediction: %0.2f'%(timePred))
print('Total time for fit and prediction: %0.2f'%(timeFit+timePred))
```
